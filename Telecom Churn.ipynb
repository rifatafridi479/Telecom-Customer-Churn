```python
df.drop('phone number', axis=1, inplace=True)
```

```python
df.columns
```

```python
value_counts = df['churn'].value_counts()

# Calculate percentages
percentages = (value_counts / len(df)) * 100

# Display results
print("Value Counts:")
print(value_counts)
print("\nPercentages:")
print(percentages)
```

```python
# Convert boolean to integer 
df['churn_int'] = df['churn'].astype(int)

# Plot pie plot for percentage of churned and retained customers

df["churn"].value_counts().plot.pie(explode=[0.005, 0.005],  autopct= "%1.1f%%", figsize= (9, 9), fontsize=20)
plt.title("Total Customer Churn Percentage", fontsize=25)
plt.show()
```

```python
# Plotting histograms for better understanding of data distributions

for i in df.select_dtypes(include="number").columns:
    sns.histplot(data=df,x=i)
    plt.show()
```

# Observation
Our histogram plots are showing that data is normally distributed which is good for data analyzing

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Convert categorical columns to numeric or exclude them
df_numeric = df.select_dtypes(include=['number'])


# Calculate the correlation matrix
correlation_matrix = df_numeric.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

# Observations:
Our heatmap show that the correlation between churn and total day minutes/ total day charge, customer service and total evening calls/ total evening charges.
